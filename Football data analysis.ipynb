```python
## number of goals in the home and away matchs for each team per each season
hgoals=sub_ginf.groupby(['ht','season'], as_index=False)['fthg'].sum()
agoals=sub_ginf.groupby(['at','season'], as_index=False)['ftag'].sum()
```

```python
goals=pd.merge(left=hgoals, right=agoals, left_on=['ht','season'], right_on=['at','season'], how='inner')
```

```python
goals['diff']=goals['fthg'] - goals['ftag']
```

```python
import scipy as sp
skw = sp.stats.skew(goals['diff'])
print("the skewness of the distribution of the difference between goals score at home and away matches = ", skw)
```

```python
sns.distplot(goals['diff'])
```

Based on this distribution the teams are more likely to score goals in home matchs than away matchs.

```python
sns.boxplot(x='season', y='diff', data=goals)
```

From these boxplots, we conclude that in each season more than 75% of the teams are more likely to score more goals in home matchs than in away matchs.
